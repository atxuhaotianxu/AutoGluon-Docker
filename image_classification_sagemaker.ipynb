# Upload data to S3

```python
import sagemaker as sage
sess = sage.Session()
prefix = 'DEMO-autogluon-example'
data_location = sess.upload_data('./data', key_prefix=prefix)
```

# Submit SageMaker Training Job

```python
import boto3

client = boto3.client('sts')
account = client.get_caller_identity()['Account']

my_session = boto3.session.Session()
region = my_session.region_name

algorithm_name = 'autogluon-example-gpu-py36-cu100'

ecr_image = '{}.dkr.ecr.{}.amazonaws.com/{}:latest'.format(account, region, algorithm_name)

print(ecr_image)
```

```python
from sagemaker.estimator import Estimator

instance_type = 'ml.p3.2xlarge'
# change it to your own account SageMaker role arn
role = "arn:aws:iam::097403188315:role/service-role/AmazonSageMaker-ExecutionRole-20200106T104790"
hyperparameters = {'epochs': 5}

estimator = Estimator(role=role,
                      train_instance_count=2,
                      train_instance_type=instance_type,
                      image_name=ecr_image,
                      hyperparameters=hyperparameters)

estimator.fit(data_location)
```
